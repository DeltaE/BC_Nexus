```python
from bcnexus.clews import model_structure as clews_const
from bcnexus import utils
```

```python
def get_storage()->dict:
    updated_resources=utils.load_config(clews_const.clews_builder_config_path)
    updated_STORAGE:dict=updated_resources['STORAGE']
    storage_techs = {}

    for resource_type,tech_info in updated_STORAGE.items():
        if resource_type=='BATTERY' :# HYDRODAM01 Handled differently
            storage_techs[resource_type] = updated_STORAGE[resource_type]

    return storage_techs
```

```python
Battery_Storages=get_storage()
Battery_Storages
```

```python
from bcnexus.clews import sets_n_ratios as snr
Years = list(range(clews_const.snapshot['start'], clews_const.snapshot['end'] + 1)) # EL
Region=clews_const.Regions
```

```python
IARList=[]
OARList=[]
```

```python
Battery_Storages
```

```python
# Create storage technologies, Fuels, Ip and Op Activity ratios
for battery in Battery_Storages:
    snr.AddActivityListItems(Years, Region, Battery_Storages[battery]['technology_from_storage'], "ELCB"+"02", IARList, g=str(Battery_Storages[battery]['mode_of_operation_charge']))
    snr.AddActivityListItems(year1=Years, region=Region, input1=Battery_Storages[battery]['technology_to_storage'], input2="ELCB"+"02", List=OARList, g='57')
```
